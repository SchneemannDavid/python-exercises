## 2. Loop Basics

#### a. While

```python
i = 5

while i <= 15:
    print(i)
```

```python

while i <= 15:
    print(i)
    i = i+1
```

```python
x = 0
while x <= 100:
    print(x)
    x = x + 2
```

```python
x = 100
while x >= -10:
    print(x)
    x = x - 5
```

```python
x = 2

while x < 1000000:
    print(x)
    x = x ** 2
```

```python
x = 100

while x > 0:
    print(x)
    x = x - 5
```

#### b. For Loops

```python
# i.

number_selection = int(input('Enter a whole number'))

for num in range(1,11):
    total = number_selection * num
    print(f"{number_selection} * {num} = {total}")
```

```python
# ii.

for num in range(10):
    print(str(num) * num)
```
