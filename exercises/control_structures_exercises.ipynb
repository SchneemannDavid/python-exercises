#### c. `break ` and   `continue`

```python
# i.

odd_input = int(input('Input odd number'))

print('The number to skip is ', odd_input)
for num in range(1, 50):
    if odd_input % 2 == 0:
        print('Please try again')
        break
    if num % 2 != 0 and num != odd_input:
        print('Here is an odd number', num)
    elif num == odd_input:
        print('Ay caramba.. Skipping this number:', odd_input)      
        continue
 
```

```python
# d. 

user_input = int(input('Please input a positive integer'))

for num in range(0,user_input+1):
    if user_input % 2 != 0:
        print('Please try again')
        break
    else:
        print(num)
    
```

```python
# e.

user_input = int(input('Please input a positive integer'))

for num in reversed(range(1, user_input+1)):
    if user_input % 2 != 0:
        print('Please try again')
        break
    else:
        print((num))
```

## 3. Fizzbuzz

```python
for num in range(1,101):
    if num % 3 == 0 and num % 5 == 0:
        print('FizzBuzz')
        continue
    elif num % 3 == 0:
        print('Fizz')
        continue
    elif num % 5 == 0:
        print('Buzz')
        continue
    else:
        print(num)
```
